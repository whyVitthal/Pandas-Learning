```python
import numpy as np
import pandas as pd 
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum() #har row mw kitni null values h bta dega
```

```python
df.isna().any() #ye btaega har column me ek null value hai ya nhi
```

Removing the missing data

```python
#remove hamesha row wise krte h jsi row me null value hogi vo reove ho jaegi
df.dropna()
```

```python
# thresh me jo value dege us hisab se check krega jsi row me utni null value hogi unko remove nhi krega else ko remove kr dega 
df
```

```python
df.dropna(thresh=3)
```

filling the mising data

```python
df
```

```python
df.fillna(300) #jaha jaha null thwa vha 300 fill ho gya
#fill column wise hota h , selection fill row wise hota h  
```

```python
values = {'A':23 , 'B':24 , 'C':35,'D':26}
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
