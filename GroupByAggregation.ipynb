```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Group by Category and calculate the sum of Sales


cat=df.groupby('Category')
cat #ye object bn jaega   jisko aap deh nhi paoge
for i , v in cat:   
    print(i) #upar left corner pr A and B alag se print hue h dekho
    print(v)

    # Basically groupby() ne jitni bhi unique categories thi (A, B, C), unke alag-alag DataFrames bana diye. Phir for i, v in cat me har iteration par i me group ka naam aaya (A, B, C) aur v me us group ka DataFrame aaya. Isliye print(i) group ka naam print karta hai aur print(v) us group ki saari rows print karta hai.
```

```python
df
```

```python
cat = df.groupby('Category')['Sales'].sum()  #sales ko select krke sum print kr diya
cat
```

```python
# Group by Store and calculate the sum of Sales
cat = df.groupby('Store')['Sales'].sum()
cat
```
