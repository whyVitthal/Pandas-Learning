```python
import numpy as np
import pandas as pd
```

Merging 2 dataframes

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries , on='employee_id' , how='inner') #inner se NaN value nhi aati and sir common aati h 
```

```python
pd.merge(employees,salaries , on='employee_id' , how='outer') #outer se sb aati h and jaha missing hota h data vha NaN
```

```python
pd.merge(employees,salaries , on='employee_id' , how='left') #employee ke basis pr merge hui h
```

```python
pd.merge(employees,salaries , on='employee_id' , how='right') #ye salaries ke basis pr hui hai
```

Concatination of 2 dataframes

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) #list me pass kro df1 and df2 and jise phle kroge vo upar aaega 
# bydefault axis = 0 hai
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df1,df2] , axis=1) #ye row wise jud gye hai
```
